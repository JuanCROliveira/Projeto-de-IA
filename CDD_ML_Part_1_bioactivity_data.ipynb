## **ChEMBL Database**

O [*ChEMBL Database*](https://www.ebi.ac.uk/chembl/) é um banco de dados que contém dados de bioatividade selecionados de mais de 2 milhões de compostos. É compilado a partir de mais de 76.000 documentos, 1,2 milhão de ensaios e os dados abrangem 13.000 alvos e 1.800 células e 33.000 indicações.

## **Pesquisar proteína-alvo**

### **Pesquisa de alvo para coronavírus**

```
# Target search for coronavirus, nesse exemplo foi utilizado o corana virus, logo aparentemente se pode utilizar qualquer um tipo de virus (lembrar do exemplo do site que ele deu como exemplo de tipo de busca no banco de dados)
target = new_client.target
target_query = target.search('coronavirus') #aqui ele da um exemplo da aromatase que e uma enzima que faz parte do citocromo p450 que é responsavel pelo cancer de mama, portanto o objetivo do trabalho e achar uma molecula/composto que seja capaz de inibir a função do aromatase, formando um farmaco a posteriori 
targets = pd.DataFrame.from_dict(target_query)
targets
```

**obs:** porque ele quis a *single protein* ?

### **Selecione e recupere dados de bioatividade para *SARS coronavirus 3C-like proteinase* (quinta entrada)**

Atribuiremos a quinta entrada (que corresponde à proteína alvo, *coronavirus 3C-like proteinase*) à variável ***selected_target***

```
selected_target = targets.target_chembl_id[4]
selected_target
```

Aqui, recuperaremos apenas dados de bioatividade para *proteínase tipo 3C do coronavírus* (CHEMBL3927) que são relatados como valores de IC$_{50}$ na unidade nM (nanomolar), lembrar  **O que significa IC50?
IC50 é o valor da concentração de um inibidor necessária para reduzir pela metade a atividade enzimática. Portanto, quanto menor for o valor de IC50, mais eficiente será o inibidor.**

```
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50") # lembrar da dose letal e dose efetiva em aulas de farmaco IC50 (A concentração inibitória média é a medida da potência de uma substância de inibir uma substância em relação a uma função biológica ou bioquímica específica.)
```

```
df = pd.DataFrame.from_dict(res)
```

Por fim, salvaremos os dados de bioatividade resultantes em um arquivo CSV **bioactivity_data.csv**.

Aparentemente, para este conjunto de dados não há dados ausentes. Mas podemos usar a célula de código acima para dados de bioatividade de outra proteína alvo.

## **Pré-processamento de dados dos dados de bioatividade**

### **Rotulagem de compostos como ativos, inativos ou intermediários**
Os dados de bioatividade estão na unidade IC50. Os compostos com valores inferiores a 1.000 nM serão considerados **ativos**, enquanto os superiores a 10.000 nM serão considerados **inativos**. Já os valores entre 1.000 e 10.000 nM serão referidos como **intermediários**.

```
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

### **Iterar o *molecule_chembl_id* para uma lista**

```
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

### **Iterar *canonical_smiles* em uma lista**

```
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

### **Iterar *standard_value* para uma lista**

```
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

### **Combine as 4 listas em um dataframe**

```
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
